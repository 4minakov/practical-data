
# Least Squares Analysis of Noisy Data

## Introduction
Processing of data containing noise in a form of outliers can be a challenging problem working with real data. This can be related for example to environmental conditions, intrument failure or human/operator errors. The outliers may significantly bias the solution leading to a wrong model. The least squares solution uses the assumption of normal distribution (introduced for convenience by C.F. Gauss in 1823) and quadratic function of residual which is sensitive/amplifies large residuals. The general approach is to 
- truncate the normal distrbution (use only data with most frequent values)
- minimize the median or absolute value of residuals 
- analysing subsets of data to obtain statistics and detect anomalous data

In this workshop, we consider only three basic methods of robust regression:

- [standard weighted least squares](#1)
- [iterative weighted least squares](#2)
- [random sample consensus (RANSAC) method](#3) 

There are many other existing methods for robust regression (e.g. Huber M-estimators, artificial neural networks etc.) and plenty of resources to learn how they work.

## Weighted Least squares
Let us revisit the problem of linear regression considere in Data_Practicals_4 in the presence of noise in a form of outliers which may bias the estimated parameters of the model.

Weighted least squares (WLS) method is an extension of ordinary least squares method that allows for the modeling of data with unequal variances. WLS assigns a weight to each data point based on the variance of its residual, and seeks to minimize the sum of these weighted squared residuals.

We aim to obtain parameters of the model with a minimum total squared error between data and predictions or minimum-norm solution assigning a weight to each data point $i$ in a form of diagonal matrix $W$ with the weights ($w_i$) on the diagonal.

The squared weighted misfit is now

$F = (x_2 - A c)^T W (x_2 - A c) \rightarrow min$

The minimum of $F$ with respect to $c$ is found by equating the derivative to zero

$\frac{\partial F}{\partial c} = 0 $

Taking the derivative, we find the solution

$ -2 A^{T} W \left(x_2 - A c\right) = 0 \\
A^{T}W x_2 - A^{T}W A c = 0 \\
c = \left(A^{T}WA\right)^{-1} A^{T}W x_2 $


This solution is unique.


### Advantages

- WLS method gives more weight to observations with smaller variances.
- Can incorporate different levels of confidence for different observations.

### Disadvantages

- Requires a good estimate of the weights for optimal performance.
- Can be sensitive to incorrect weight specification.

Weighted Least Squares can be a powerful tool for regression analysis when the assumptions of ordinary least squares are not met, particularly when the data contain outliers.


```python
import numpy as np
x1 = np.array([-0.5, -0.1, 0.2, 0.5])
x2 = np.array([-0.4, -0.1, 0.2, 0.9])
x2std = np.array([0.1, 0.1, 0.1, 0.1]) #standard error
# weights (inverse of standard errors)
weights = np.array(x2std)**(-1)
# Create the matrix A by concatenating x1 with an array of ones
A = np.column_stack((x1, np.ones_like(x1)))    
# Weighted A and x2
A_weighted = np.sqrt(weights)[:, np.newaxis] * A
x2_weighted = np.sqrt(weights) * x2   
# Minimum norm solution
c = np.linalg.inv(A_weighted.T @ A_weighted) @ A_weighted.T @ x2_weighted
residuals = x2 - (A @ c) 
# Covariance (Optional, you can update this part based on your needs)
cv = np.linalg.inv(A.T @ np.diag(1/np.array(x2std)**2) @ A) 
print("Coefficients:", c)
print("Covariance:", cv)
print("Weights:", weights)
print("Residuals:", residuals)
# Prediction of data based on estimated parameters
p1 = np.linspace(-1, 1, 10)
xp = p1 * c[0] + c[1]
#Plotting
plt.figure(1)
plt.clf()
plt.plot(p1,xp,'r')
plt.plot(x1, x2, 'ok')
plt.errorbar(x1, x2, yerr=x2std, fmt='.k')
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```

## Iterative weighted least squares

1. **Initial Weights**: An initial model is fit using ordinary least squares method to estimate the residuals and their variances, which are then used to calculate initial weights.
2. **Iterate**: Fit the WLS model using the weights, update the weights based on new residuals, and refit the model until convergence.


```python
import numpy as np


x1 = np.array([-0.5, -0.1, 0.2, 0.5])
x2 = np.array([-0.4, -0.1, 0.2, 0.9])

x2std = np.array([0.1, 0.1, 0.1, 0.5]) #standard error


# Initial weights (could be all ones if you don't have a better estimate)
weights = np.array(x2std)**(-1)

# Iterative procedure
max_iter = 100  # Maximum number of iterations
tolerance = 1e-6  # Convergence tolerance
prev_c = np.zeros(2)  # Initialize previous coefficients to zeros

for i in range(max_iter):
    # Create the matrix A by concatenating x1 with an array of ones
    A = np.column_stack((x1, np.ones_like(x1)))
    
    # Weighted A and x2
    A_weighted = np.sqrt(weights)[:, np.newaxis] * A
    x2_weighted = np.sqrt(weights) * x2
    
    # Minimum norm solution
    c = np.linalg.inv(A_weighted.T @ A_weighted) @ A_weighted.T @ x2_weighted
    
    # Check for convergence
    if np.all(np.abs(c - prev_c) < tolerance):
        print("Converged.")
        break

    # Update weights based on new residuals
    residuals = x2 - (A @ c) 
    weights = 1 / (residuals ** 2 + 1e-5) # least squares norm
    
    # Update previous coefficients
    prev_c = c

# Covariance (Optional, you can update this part based on your needs)
cv = np.linalg.inv(A.T @ np.diag(1/np.array(x2std)**2) @ A) 

print("Coefficients:", c)
print("Covariance:", cv)
print("Weights:", weights)
print("Residuals:", residuals)

# Estimation
p1 = np.linspace(-1, 1, 10)
xp = p1 * c[0] + c[1]

plt.figure(1)
plt.clf()
plt.plot(p1,xp,'r')
plt.plot(x1, x2, 'ok')
plt.errorbar(x1, x2, yerr=x2std, fmt='.k')
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```

## Random Sample Consensus (RanSaC) method

## Algorithm Steps
1. **Random Sample**
    - Randomly select a subset of the data.
  
2. **Fit Model**
    - Fit a model to the selected subset.
  
3. **Inliers**
    - Find all data that fits the model within a certain tolerance.
  
4. **Re-fit Model**
    - Re-fit the model using all identified inliers.
  
5. **Evaluate Model**
    - Evaluate the model based on the number of inliers.
  
6. **Iterate**
    - Repeat the above steps for a fixed number of iterations or until a satisfactory model is found.
    
    ## Parameters
- **Number of Iterations**: How many subsets to try.
- **Tolerance**: How close a point must be to the model to be considered an inlier.
- **Minimum Inliers**: Minimum number of inliers required to consider a model good.

# Conclusions

The robust methods allows to reduce the effect of outliers in estimation of model parameters in noisy data.
Each of three considered methods has advantages and drawbacks. Such as often in practice we know little about actual uncertainties of the data points required for the weighted least squares method. The assumption of the dependence of data weight on its residual does not necessarily holds. In RANSAC method the threshold and number of trial are tuning parameters that need to be chosen. Similar dependence on tuning parameters applies to other methods of robust regression. 
Overall although the robust regression methods (with some experience) can help to minimize the impact of large errors in data, the best approach would be to avoid the outliers on the stage of data aquisition and quality control.
