# import Python libraries

```python
import numpy as np # numpy does general array operations

import matplotlib.pyplot as plt # matplotlib does plotting
import matplotlib.ticker as ticker

import tensorflow as tf  # tensorflow is a python library to use neural networks
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
from tensorflow.keras.optimizers import *

from sklearn.model_selection import train_test_split # scikit learn for preparing and measuring ML data
from sklearn.metrics import accuracy_score
```

# load data: pockmark images, and labels: yes (1) or no (0) pockmark

```python
data = np.load('data_gn1101.npy')
print('The data shape is: (# of images, # of pixels X, # of pixels Y) ', data.shape)
labels = np.load('labels_gn1101.npy')
print('The number of labels is: ', labels.shape[0])
```

```python
# count how many 1 labels (pockmarks) are in the data
count_of_pockmarks = np.count_nonzero(labels == 1)
print('There are ' + str(count_of_pockmarks) + ' pockmarks in the data, and ' + str(labels.shape[0]-count_of_pockmarks) + ' non pockmarks')
```

# split the data into training and testing

```python
# shuffle the data around! (data is initially unsorted with pockmarks first)
# test size = what percentage of the data to leave out for testing
# random state = set to any given number to remember the randomization for next time

xtrain, xtest, ytrain, ytest = train_test_split(data, labels, test_size=0.15, shuffle = True, random_state=40)

# xtrain = training images
# xtest = testing images
# ytrain = training labels
# ytest = testing labels
```

# design the CNN model

```python
image_x = 20
image_y = 20

def CNN(n_conv_filters, n_dense_nodes, k_size, input_size=[image_x, image_y, 1]):
    
    inputs = Input(input_size)
    
    layer1 = Conv2D(n_conv_filters, kernel_size=(k_size, k_size), activation='relu')(inputs) \
    
    layer2 = MaxPooling2D((2, 2))(layer1)
    
    layer3 = Flatten()(layer2)
    
    layer4 = Dense(n_dense_nodes, activation = 'relu')(layer3)
    
    layer5 = Dense(1, activation = 'sigmoid')(layer4)

    model = Model(inputs=[inputs], outputs=[layer5])
    model.summary()
    
    return model
```

```python
n_conv_filters = 32
n_dense_nodes = 100
k_size = 3

model_CNN = CNN(n_conv_filters, n_dense_nodes, k_size, input_size=[image_x, image_y, 1])
```

# compile the model and start training

```python
model_CNN.compile(optimizer=Adam(learning_rate=1e-5), loss='binary_crossentropy', metrics=['accuracy'])


training_history_CNN = model_CNN.fit(xtrain, ytrain, batch_size=32, epochs=400)
```

# plot the history of its training (accuracy / binary cross entropy over time)

```python
fig,ax = plt.subplots(figsize=(8,6))
ax.plot(training_history_CNN.history['loss'])
ax.set_ylabel('Binary Cross Entropy')
ax.set_xlabel('Epochs')
ax.set_title('Training Loss for CNN')
```

```python
fig,ax = plt.subplots(figsize=(8,6))
ax.plot(training_history_CNN.history['accuracy'])
ax.set_ylabel('Accuracy')
ax.set_xlabel('Epochs')
ax.set_title('Pockmark ID Accuracy With Training Time')
```

# predict on the unseen testing images (xtest)

```python
ypred = model_CNN.predict(xtest)
```

```python
num = 74

plt.imshow(xtest[num,:,:], aspect='auto', cmap='Greys', vmin=0, vmax=0.07)
print('The predicted probability is: ', ypred[num][0])
print('Rounding this gives: ', round(ypred[num][0]))
print('')
print('The actual classification is: ', ytest[num])
```

```python
# running this calculates the average BCE and accuracy for all the testing data

test_bce, test_accuracy = model_CNN.evaluate(xtest, ytest, verbose=0)
```
