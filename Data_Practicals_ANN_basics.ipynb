- What is linear regression?

## Basic Concepts

- Loss function

- Gradient descent.

Gradient descent is an optimization algorithm used to minimize some function by iteratively moving towards the minimum of that function. The function we want to minimize in the context of machine learning is usually a loss function $L(\theta)$, which measures how well our model fits the data.

The gradient descent algorithm updates the parameters $\theta$ of the model by following the negative of the gradient of the loss function. Here's the basic update rule:

$$
\theta_{\text{new}} = \theta_{\text{old}} - \eta \cdot \nabla_\theta L(\theta)
$$

where:
- $\theta$ represents the parameters of our model.
- $\eta$ is the learning rate, a hyperparameter that determines the step size during the update.
- $\nabla_\theta L(\theta)$ is the gradient of the loss function with respect to the parameters $\theta$.

The gradient $\nabla_\theta L(\theta)$ points in the direction of the steepest ascent of the loss function. Since we want to minimize the loss, we move in the opposite direction.

For a loss function $L$ that is a function of a single parameter $\theta$, the derivative $\frac{dL}{d\theta}$ gives the slope of the function at that point. If the derivative is positive, the function is increasing, and if the derivative is negative, the function is decreasing. Therefore, by subtracting a fraction of the derivative from $\theta$, we move towards the minimum of the function.

The mean squared error (MSE) loss function for linear regression is defined as:

$$
L(\theta) = \frac{1}{n} \sum_{i=1}^{n} (y_i - (mx_i + b))^2
$$

The partial derivatives of the MSE loss function with respect to the model parameters $m$ and $b$ are:

$$
\frac{\partial L}{\partial m} = \frac{-2}{n} \sum_{i=1}^{n} x_i(y_i - (mx_i + b))
$$

$$
\frac{\partial L}{\partial b} = \frac{-2}{n} \sum_{i=1}^{n} (y_i - (mx_i + b))
$$

The update rules for $m$ and $b$ will be:

$$
m_{\text{new}} = m_{\text{old}} - \eta \cdot \frac{\partial L}{\partial m}
$$

$$
b_{\text{new}} = b_{\text{old}} - \eta \cdot \frac{\partial L}{\partial b}
$$

In each iteration of gradient descent, we compute these derivatives to find the gradient and then update $m$ and $b$ accordingly until we reach convergence, meaning that the updates are very small or stop changing.



## Algorithm Steps

1. **Initialize Parameters**
    - Set the number of iterations and learning rate.
    - Initialize weights and biases randomly.


2. **Training Loop**
    - Forward pass to make predictions.
    - Compute loss
    - Backward pass to update weights and biases.
  
  
3. **Data inference**
    - Apply obtained weights and biases to predict data according to the assumed linear model 

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of training iterations (epochs)
numEpoch = 20

#  Learning rate
learningRate = 0.1

# Initialize weights and biases
w = np.random.randn()
b = np.random.randn()

# data
x, y = np.array([1, 2, 3, 4]), np.array([2, 4, 6, 8])

# Training loop
# The goal is to find optimal weights and biases
# y = weight * x + bias
# which minimize the misfit expressed as 

loss = np.zeros(numEpoch) #initialize objective function 

for i in range(numEpoch):
    # Linear model prediction
    y_it = w * x + b

    # Calculate objective function (loss)
    loss[i] = np.mean((y_it - y)**2)

    # Gradient of objective function wrt estimated parameters
    dw = np.mean(2 * (y_it - y) * x)
    db = np.mean(2 * (y_it - y))

    # Update weights
    w = w - learningRate * dw
    b = b - learningRate * db
    
plt.plot(loss,'o-')   
plt.xlabel('Epoch')
plt.ylabel('Average of Training error')
plt.show()
```

## Results

- Make prediction of linear model based on obtained weights 
- Plot the data points and the fitted line.

```python
# Prediction of data

x_pred = np.linspace(np.min(x), np.max(x), 100)
y_pred = w * x_pred + b

# Display pedicted linear model
print(f"Predicted linear model: y = {w} * x + {b}")

# Plot the data points and the fitted line
plt.scatter(x, y, c='red', marker='o', label='Data Points')
plt.plot(x_pred, y_pred, 'b-', linewidth=2, label='Fitted Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Line Fitting using ANN')
plt.legend()
plt.grid(True)
plt.show()
        
```

## Conclusion
- The objective of training an artificial neural network is to find weights that can be applied to the input data to obtain desired output
- The weights are sought by minimizing a misfit (loss) function in the direction of its maximum change
- The main power of ANN that it can be applied to find a complex non-linear relationship between input and output. In this case, a more complex objective functions and multiple layers of neurons must be used.
